```python
def menu():
    print ("""Operações com frações:
             
              (1) Criar uma fração;
              (2) Simplificar uma fração;
              (3) Somar;
              (4) Subtrair;
              (5) Multiplicar;
              (6) Dividir
              (7) Gerar lista com n frações;
              (8) Somar lista;
              (9) Maior fração;
              (10)Menor fração;
              (11)Maior Fração de uma lista;
              (12)Menor Fração de uma lista;
              
              (0) Sair""")


def criar_fracao(numerador, denominador):
    return (numerador, denominador)

def ver_fracao(f):
    print ("%.0f" % f[0] +"/"+ "%.0f" % f[1])

def mdc(a,b):
    while b!=0:
        resto=a % b
        a=b
        b=resto

    return a

def simplificar_fracao(f):
    num,denom=f
    a=mdc(num,denom)
    return (num/a,denom/a)

def som_fracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simplificar_fracao(criar_fracao((n1*d1+n2*d2),d1*d2))

def sub_fracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simplificar_fracao(criar_fracao((n1*d1-n2*d2),d1*d2))

def mult_fracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simplificar_fracao(criar_fracao((n1*n2),(d1*d2)))

def div_fracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simplificar_fracao(criar_fracao((n1*d2),(d1*n2)))

def gerar_lista(n):
    import random
    lista_frac=[]
    for i in range(n):
        num=random.randint (0,100)
        denom=random.randint (1,100)
        f=criar_fracao(num,denom)
        lista_frac.append (f)
    return lista_frac
    
    
def somar_lista(lista_frac):
    ac=(0,1)
    for elem in range(len(lista_frac)):
        ac=som_fracao(ac,lista_frac[elem])
    return ac

def maior_fracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    n3=n1*d2
    n4=n2*d1
    if n3>n4:
        return f1
    else:
        return f2

def menor_fracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    n3=n1*d2
    n4=n2*d1
    if n3<n4:
        return f1
    else:
        return f2

def maior_fracao_lis(lista_frac):
    maior=(0,0)
    for elem in lista_frac:
        maior=maior_fracao(maior,elem)
    return maior
        
def menor_fracao_lis(lista_frac):
    menor=lista_frac[0]
    for elem in lista_frac:
        menor=menor_fracao(menor,elem)
    return menor

import re
lis_gerada=[]
op="1"
while (op!="0"):
    menu()
    op=input("Qual é a sua opção:")
    if (op=="1"):
        a=""
        b=""
        i=0
        while (re.match("^[a-z]*$", a) or re.match("^[a-z]*$", b) or a=="" or b==""):
            if(i==1):
                print("Numerador ou denominador errados!")
            print ("Introduza o numerador e denominador para a primeira fração.")
            a=input("Numerador: ")
            b=input("Denominador: ")
            i=1
        f1=simplificar_fracao(criar_fracao(int(a),int(b)))
        ver_fracao(f1)
    elif (op=="2"):
        fracao=simplificar_fracao(f1)
        ver_fracao(fracao)
   
    elif (op=="3"):
        a=""
        b=""
        c=""
        d=""
        i=0
        while (re.match("^[a-z]*$", a) or re.match("^[a-z]*$", b) or a=="" or b==""):
            if(i==1):
                print("Numerador ou denominador errados!")
            print ("Introduza o numerador e denominador para a primeira fração.")
            a=input("Numerador: ")
            b=input("Denominador: ")
            i=1
        f1=simplificar_fracao(criar_fracao(int(a),int(b)))
        i=0
        while (re.match("^[a-z]*$", c) or re.match("^[a-z]*$", d) or c=="" or d==""):
            if(i==1):
                print("Numerador ou denominador errados!")
            print ("Introduza o numerador e denominador para a segunda fração.")
            c=input("Numerador: ")
            d=input("Denominador: ")
            i=1
        f2=simplificar_fracao(criar_fracao(int(c),int(d)))
        soma=som_fracao (f1,f2)
        ver_fracao(soma)
    
    elif (op=="4"):
        a=""
        b=""
        c=""
        d=""
        i=0
        while (re.match("^[a-z]*$", a) or re.match("^[a-z]*$", b) or a=="" or b==""):
            if(i==1):
                print("Numerador ou denominador errados!")
            print ("Introduza o numerador e denominador para a primeira fração.")
            a=input("Numerador: ")
            b=input("Denominador: ")
            i=1
        f1=simplificar_fracao(criar_fracao(int(a),int(b)))
        i=0
        while (re.match("^[a-z]*$", c) or re.match("^[a-z]*$", d) or c=="" or d==""):
            if(i==1):
                print("Numerador ou denominador errados!")
            print ("Introduza o numerador e denominador para a segunda fração.")
            c=input("Numerador: ")
            d=input("Denominador: ")
            i=1
        f2=simplificar_fracao(criar_fracao(int(c),int(d)))
        subtracao=sub_fracao(f1,f2)
        ver_fracao(subtracao)
    
    elif (op=="5"):
        a=""
        b=""
        c=""
        d=""
        i=0
        while (re.match("^[a-z]*$", a) or re.match("^[a-z]*$", b) or a=="" or b==""):
            if(i==1):
                print("Numerador ou denominador errados!")
            print ("Introduza o numerador e denominador para a primeira fração.")
            a=input("Numerador: ")
            b=input("Denominador: ")
            i=1
        f1=simplificar_fracao(criar_fracao(int(a),int(b)))
        i=0
        while (re.match("^[a-z]*$", c) or re.match("^[a-z]*$", d) or c=="" or d==""):
            if(i==1):
                print("Numerador ou denominador errados!")
            print ("Introduza o numerador e denominador para a segunda fração.")
            c=input("Numerador: ")
            d=input("Denominador: ")
            i=1
        f2=simplificar_fracao(criar_fracao(int(c),int(d)))
        multiplicacao=mult_fracao (f1,f2)
        ver_fracao(multiplicacao)
    
    elif (op=="6"):
        a=""
        b=""
        c=""
        d=""
        i=0
        while (re.match("^[a-z]*$", a) or re.match("^[a-z]*$", b) or a=="" or b==""):
            if(i==1):
                print("Numerador ou denominador errados!")
            print ("Introduza o numerador e denominador para a primeira fração.")
            a=input("Numerador: ")
            b=input("Denominador: ")
            i=1
        f1=simplificar_fracao(criar_fracao(int(a),int(b)))
        i=0
        while (re.match("^[a-z]*$", c) or re.match("^[a-z]*$", d) or c=="" or d==""):
            if(i==1):
                print("Numerador ou denominador errados!")
            print ("Introduza o numerador e denominador para a segunda fração.")
            c=input("Numerador: ")
            d=input("Denominador: ")
            i=1
        f2=simplificar_fracao(criar_fracao(int(c),int(d)))
        divisao=div_fracao (f1,f2)
        ver_fracao(divisao)
        
    elif (op=="7"):
        n=""
        i=0
        while (re.match("^[a-z]*$", n) or n=="" or int(n)<1):
            if(i==1):
                print("Erro na introdução do número de elementos da lista!")
            n=input("Introduza o número de elementos da lista: ")
            i=1
        lis_gerada=gerar_lista(int(n))
        viz_lis=[]
        for index in range(len(lis_gerada)):
            num, denom = lis_gerada[index]
            viz_lis.append(str(num)+"/"+str(denom))
        print(viz_lis)
    
    elif (op=="8" and len(lis_gerada)>0):
        soma_lis=somar_lista(lis_gerada)
        ver_fracao(soma_lis)
    elif (op=="9"):
        a=""
        b=""
        c=""
        d=""
        i=0
        while (re.match("^[a-z]*$", a) or re.match("^[a-z]*$", b) or a=="" or b==""):
            if(i==1):
                print("Numerador ou denominador errados!")
            print ("Introduza o numerador e denominador para a primeira fração.")
            a=input("Numerador: ")
            b=input("Denominador: ")
            i=1
        f1=simplificar_fracao(criar_fracao(int(a),int(b)))
        i=0
        while (re.match("^[a-z]*$", c) or re.match("^[a-z]*$", d) or c=="" or d==""):
            if(i==1):
                print("Numerador ou denominador errados!")
            print ("Introduza o numerador e denominador para a segunda fração.")
            c=input("Numerador: ")
            d=input("Denominador: ")
            i=1
        f2=simplificar_fracao(criar_fracao(int(c),int(d)))
        maior=maior_fracao(f1,f2)
        print(maior)
    elif (op=="10"):
        a=""
        b=""
        c=""
        d=""
        i=0
        while (re.match("^[a-z]*$", a) or re.match("^[a-z]*$", b) or a=="" or b==""):
            if(i==1):
                print("Numerador ou denominador errados!")
            print ("Introduza o numerador e denominador para a primeira fração.")
            a=input("Numerador: ")
            b=input("Denominador: ")
            i=1
        f1=simplificar_fracao(criar_fracao(int(a),int(b)))
        i=0
        while (re.match("^[a-z]*$", c) or re.match("^[a-z]*$", d) or c=="" or d==""):
            if(i==1):
                print("Numerador ou denominador errados!")
            print ("Introduza o numerador e denominador para a segunda fração.")
            c=input("Numerador: ")
            d=input("Denominador: ")
            i=1
        f2=simplificar_fracao(criar_fracao(int(c),int(d)))
        menor=menor_fracao(f1,f2)
        print(menor)
    elif (op=="11" and len(lis_gerada)>0):
        maifl=maior_fracao_lis(lis_gerada)
        print(maifl)
    elif (op=="12" and len(lis_gerada)>0):
        menfl=menor_fracao_lis(lis_gerada)
        print(menfl)
    elif (op=="0"):
        print("Até à próxima!")
    else: 
        print("Opção inválida!")
```
